Tinting gray-scale images<br>
=========================<br>
It can be useful to artificially tint an image with some color, either to<br>
highlight particular regions of an image or maybe just to liven up a grayscale<br>
image. This example demonstrates image-tinting by scaling RGB values and by<br>
adjusting colors in the HSV color-space.<br>
In 2D, color images are often represented in RGB---3 layers of 2D arrays, where<br>
the 3 layers represent (R)ed, (G)reen and (B)lue channels of the image. The<br>
simplest way of getting a tinted image is to set each RGB channel to the<br>
grayscale image scaled by a different multiplier for each channel. For example,<br>
multiplying the green and blue channels by 0 leaves only the red channel and<br>
produces a bright red image. Similarly, zeroing-out the blue channel leaves<br>
only the red and green channels, which combine to form yellow.<br>


```python
import matplotlib.pyplot as plt
from skimage import data
from skimage import color
from skimage import img_as_float
```

```python
grayscale_image = img_as_float(data.camera()[::2, ::2])
image = color.gray2rgb(grayscale_image)
```

```python
red_multiplier = [1, 0, 0]
yellow_multiplier = [1, 1, 0]
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4),
                               sharex=True, sharey=True)
ax1.imshow(red_multiplier * image)
ax2.imshow(yellow_multiplier * image)
```

####################################################################<br>
In many cases, dealing with RGB values may not be ideal. Because of that,<br>
there are many other `color spaces`_ in which you can represent a color<br>
image. One popular color space is called HSV, which represents hue (~the<br>
color), saturation (~colorfulness), and value (~brightness). For example, a<br>
color (hue) might be green, but its saturation is how intense that green is<br>
---where olive is on the low end and neon on the high end.<br>
<br>
In some implementations, the hue in HSV goes from 0 to 360, since hues wrap<br>
around in a circle. In scikit-image, however, hues are float values from 0<br>
to 1, so that hue, saturation, and value all share the same scale.<br>
<br>
.. _color spaces:<br>
    https://en.wikipedia.org/wiki/List_of_color_spaces_and_their_uses<br>
<br>
Below, we plot a linear gradient in the hue, with the saturation and value<br>
turned all the way up:

```python
import numpy as np
```

```python
hue_gradient = np.linspace(0, 1)
hsv = np.ones(shape=(1, len(hue_gradient), 3), dtype=float)
hsv[:, :, 0] = hue_gradient
```

```python
all_hues = color.hsv2rgb(hsv)
```

```python
fig, ax = plt.subplots(figsize=(5, 2))
# Set image extent so hues go from 0 to 1 and the image is a nice aspect ratio.
ax.imshow(all_hues, extent=(0 - 0.5 / len(hue_gradient),
                            1 + 0.5 / len(hue_gradient), 0, 0.2))
ax.set_axis_off()
```

####################################################################<br>
Notice how the colors at the far left and far right are the same. That<br>
reflects the fact that the hues wrap around like the color wheel (see HSV_<br>
for more info).<br>
<br>
.. _HSV: https://en.wikipedia.org/wiki/HSL_and_HSV<br>
<br>
Now, let's create a little utility function to take an RGB image and:<br>
<br>
1. Transform the RGB image to HSV 2. Set the hue and saturation 3.<br>
Transform the HSV image back to RGB

```python
def colorize(image, hue, saturation=1):
    """ Add color of the given hue to an RGB image.
    By default, set the saturation to 1 so that the colors pop!
    """
    hsv = color.rgb2hsv(image)
    hsv[:, :, 1] = saturation
    hsv[:, :, 0] = hue
    return color.hsv2rgb(hsv)
```

####################################################################<br>
Notice that we need to bump up the saturation; images with zero saturation<br>
are grayscale, so we need to a non-zero value to actually see the color<br>
we've set.<br>
<br>
Using the function above, we plot six images with a linear gradient in the<br>
hue and a non-zero saturation:

```python
hue_rotations = np.linspace(0, 1, 6)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, sharex=True, sharey=True)
```

```python
for ax, hue in zip(axes.flat, hue_rotations):
    # Turn down the saturation to give it that vintage look.
    tinted_image = colorize(image, hue, saturation=0.3)
    ax.imshow(tinted_image, vmin=0, vmax=1)
    ax.set_axis_off()
fig.tight_layout()
```

####################################################################<br>
You can combine this tinting effect with numpy slicing and fancy-indexing<br>
to selectively tint your images. In the example below, we set the hue of<br>
some rectangles using slicing and scale the RGB values of some pixels found<br>
by thresholding. In practice, you might want to define a region for tinting<br>
based on segmentation results or blob detection methods.

```python
from skimage.filters import rank
```

Square regions defined as slices over the first two dimensions.

```python
top_left = (slice(25),) * 2
bottom_right = (slice(-25, None),) * 2
```

```python
sliced_image = image.copy()
sliced_image[top_left] = colorize(image[top_left], 0.82, saturation=0.5)
sliced_image[bottom_right] = colorize(image[bottom_right], 0.5, saturation=0.5)
```

Create a mask selecting regions with interesting texture.

```python
noisy = rank.entropy(grayscale_image, np.ones((9, 9)))
textured_regions = noisy > 4.25
# Note that using `colorize` here is a bit more difficult, since `rgb2hsv`
# expects an RGB image (height x width x channel), but fancy-indexing returns
# a set of RGB pixels (# pixels x channel).
masked_image = image.copy()
masked_image[textured_regions, :] *= red_multiplier
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(8, 4),
                               sharex=True, sharey=True)
ax1.imshow(sliced_image)
ax2.imshow(masked_image)
```
