Histogram Equalization<br>
======================<br>
This examples enhances an image with low contrast, using a method called<br>
*histogram equalization*, which "spreads out the most frequent intensity<br>
values" in an image [1]_. The equalized image has a roughly linear cumulative<br>
distribution function.<br>
While histogram equalization has the advantage that it requires no parameters,<br>
it sometimes yields unnatural looking images.  An alternative method is<br>
*contrast stretching*, where the image is rescaled to include all intensities<br>
that fall within the 2nd and 98th percentiles [2]_.<br>
.. [1] https://en.wikipedia.org/wiki/Histogram_equalization<br>
.. [2] http://homepages.inf.ed.ac.uk/rbf/HIPR2/stretch.htm<br>


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
from skimage import data, img_as_float
from skimage import exposure
```

```python
matplotlib.rcParams['font.size'] = 8
```

```python
def plot_img_and_hist(image, axes, bins=256):
    """Plot an image along with its histogram and cumulative histogram.
    """
    image = img_as_float(image)
    ax_img, ax_hist = axes
    ax_cdf = ax_hist.twinx()

    # Display image
    ax_img.imshow(image, cmap=plt.cm.gray)
    ax_img.set_axis_off()

    # Display histogram
    ax_hist.hist(image.ravel(), bins=bins, histtype='step', color='black')
    ax_hist.ticklabel_format(axis='y', style='scientific', scilimits=(0, 0))
    ax_hist.set_xlabel('Pixel intensity')
    ax_hist.set_xlim(0, 1)
    ax_hist.set_yticks([])

    # Display cumulative distribution
    img_cdf, bins = exposure.cumulative_distribution(image, bins)
    ax_cdf.plot(bins, img_cdf, 'r')
    ax_cdf.set_yticks([])
    return ax_img, ax_hist, ax_cdf
```

Load an example image

```python
img = data.moon()
```

Contrast stretching

```python
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))
```

Equalization

```python
img_eq = exposure.equalize_hist(img)
```

Adaptive Equalization

```python
img_adapteq = exposure.equalize_adapthist(img, clip_limit=0.03)
```

Display results

```python
fig = plt.figure(figsize=(8, 5))
axes = np.zeros((2, 4), dtype=np.object)
axes[0, 0] = fig.add_subplot(2, 4, 1)
for i in range(1, 4):
    axes[0, i] = fig.add_subplot(2, 4, 1+i, sharex=axes[0,0], sharey=axes[0,0])
for i in range(0, 4):
    axes[1, i] = fig.add_subplot(2, 4, 5+i)
```

```python
ax_img, ax_hist, ax_cdf = plot_img_and_hist(img, axes[:, 0])
ax_img.set_title('Low contrast image')
```

```python
y_min, y_max = ax_hist.get_ylim()
ax_hist.set_ylabel('Number of pixels')
ax_hist.set_yticks(np.linspace(0, y_max, 5))
```

```python
ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_rescale, axes[:, 1])
ax_img.set_title('Contrast stretching')
```

```python
ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_eq, axes[:, 2])
ax_img.set_title('Histogram equalization')
```

```python
ax_img, ax_hist, ax_cdf = plot_img_and_hist(img_adapteq, axes[:, 3])
ax_img.set_title('Adaptive equalization')
```

```python
ax_cdf.set_ylabel('Fraction of total intensity')
ax_cdf.set_yticks(np.linspace(0, 1, 5))
```
