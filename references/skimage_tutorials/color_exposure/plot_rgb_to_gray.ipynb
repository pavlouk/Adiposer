================<br>
RGB to grayscale<br>
================<br>
This example converts an image with RGB channels into an image with a single<br>
grayscale channel.<br>
The value of each grayscale pixel is calculated as the weighted sum of the<br>
corresponding red, green and blue pixels as::<br>
        Y = 0.2125 R + 0.7154 G + 0.0721 B<br>
These weights are used by CRT phosphors as they better represent human<br>
perception of red, green and blue than equal weights. [1]_<br>
References<br>
----------<br>
.. [1] http://poynton.ca/PDFs/ColorFAQ.pdf<br>


```python
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage.color import rgb2gray
```

```python
original = data.astronaut()
grayscale = rgb2gray(original)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
```

```python
ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")
```

```python
fig.tight_layout()
plt.show()
```
