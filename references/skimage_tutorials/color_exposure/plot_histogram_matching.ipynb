Histogram matching<br>
==================<br>
This example demonstrates the feature of histogram matching. It manipulates the<br>
pixels of an input image so that its histogram matches the histogram of the<br>
reference image. If the images have multiple channels, the matching is done<br>
independently for each channel, as long as the number of channels is equal in<br>
the input image and the reference.<br>
Histogram matching can be used as a lightweight normalisation for image<br>
processing, such as feature matching, especially in circumstances where the<br>
images have been taken from different sources or in different conditions (i.e.<br>
lighting).<br>


```python
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage import exposure
from skimage.exposure import match_histograms
```

```python
reference = data.coffee()
image = data.chelsea()
```

```python
matched = match_histograms(image, reference, channel_axis=-1)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()
```

```python
ax1.imshow(image)
ax1.set_title('Source')
ax2.imshow(reference)
ax2.set_title('Reference')
ax3.imshow(matched)
ax3.set_title('Matched')
```

```python
plt.tight_layout()
plt.show()
```

####################################################################<br>
To illustrate the effect of the histogram matching, we plot for each<br>
RGB channel, the histogram and the cumulative histogram. Clearly,<br>
the matched image has the same cumulative histogram as the reference<br>
image for each channel.

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(8, 8))
```

```python
for i, img in enumerate((image, reference, matched)):
    for c, c_color in enumerate(('red', 'green', 'blue')):
        img_hist, bins = exposure.histogram(img[..., c], source_range='dtype')
        axes[c, i].plot(bins, img_hist / img_hist.max())
        img_cdf, bins = exposure.cumulative_distribution(img[..., c])
        axes[c, i].plot(bins, img_cdf)
        axes[c, 0].set_ylabel(c_color)
```

```python
axes[0, 0].set_title('Source')
axes[0, 1].set_title('Reference')
axes[0, 2].set_title('Matched')
```

```python
plt.tight_layout()
plt.show()
```
