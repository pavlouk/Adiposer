Label image regions<br>
===================<br>
This example shows how to segment an image with image labelling. The following<br>
steps are applied:<br>
1. Thresholding with automatic Otsu method<br>
2. Close small holes with binary closing<br>
3. Remove artifacts touching image border<br>
4. Measure image regions to filter small objects<br>


```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
from skimage import data
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb
```

```python
image = data.coins()[50:-50, 50:-50]
```

apply threshold

```python
thresh = threshold_otsu(image)
bw = closing(image > thresh, square(3))
```

remove artifacts connected to image border

```python
cleared = clear_border(bw)
```

label image regions

```python
label_image = label(cleared)
# to make the background transparent, pass the value of `bg_label`,
# and leave `bg_color` as `None` and `kind` as `overlay`
image_label_overlay = label2rgb(label_image, image=image, bg_label=0)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(image_label_overlay)
```

```python
for region in regionprops(label_image):
    # take regions with large enough areas
    if region.area >= 100:
        # draw rectangle around segmented coins
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                                  fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)
```

```python
ax.set_axis_off()
plt.tight_layout()
plt.show()
```
