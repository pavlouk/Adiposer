==============<br>
Normalized Cut<br>
==============<br>
This example constructs a Region Adjacency Graph (RAG) and recursively performs<br>
a Normalized Cut on it [1]_.<br>
References<br>
----------<br>
.. [1] Shi, J.; Malik, J., "Normalized cuts and image segmentation",<br>
       Pattern Analysis and Machine Intelligence,<br>
       IEEE Transactions on, vol. 22, no. 8, pp. 888-905, August 2000.<br>


```python
from skimage import data, segmentation, color
from skimage.future import graph
from matplotlib import pyplot as plt
```

```python
img = data.coffee()
```

```python
labels1 = segmentation.slic(img, compactness=30, n_segments=400,
                            start_label=1)
out1 = color.label2rgb(labels1, img, kind='avg', bg_label=0)
```

```python
g = graph.rag_mean_color(img, labels1, mode='similarity')
labels2 = graph.cut_normalized(labels1, g)
out2 = color.label2rgb(labels2, img, kind='avg', bg_label=0)
```

```python
fig, ax = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(6, 8))
```

```python
ax[0].imshow(out1)
ax[1].imshow(out2)
```

```python
for a in ax:
    a.axis('off')
```
