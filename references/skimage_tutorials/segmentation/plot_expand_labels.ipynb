Expand segmentation labels without overlap<br>
==========================================<br>
Given several connected components represented by a label image, these<br>
connected components can be expanded into background regions using<br>
:py:func:`skimage.segmentation.expand_labels`.<br>
In contrast to :py:func:`skimage.morphology.dilation` this method will<br>
not let connected components expand into neighboring connected components<br>
with lower label number.<br>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage.filters import sobel
from skimage.measure import label
from skimage.segmentation import watershed, expand_labels
from skimage.color import label2rgb
from skimage import data
```

```python
coins = data.coins()
```

Make segmentation using edge-detection and watershed.

```python
edges = sobel(coins)
```

Identify some background and foreground pixels from the intensity values.<br>
These pixels are used as seeds for watershed.

```python
markers = np.zeros_like(coins)
foreground, background = 1, 2
markers[coins < 30.0] = background
markers[coins > 150.0] = foreground
```

```python
ws = watershed(edges, markers)
seg1 = label(ws == foreground)
```

```python
expanded = expand_labels(seg1, distance=10)
```

Show the segmentations.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 5),
                         sharex=True, sharey=True)
```

```python
color1 = label2rgb(seg1, image=coins, bg_label=0)
axes[0].imshow(color1)
axes[0].set_title('Sobel+Watershed')
```

```python
color2 = label2rgb(expanded, image=coins, bg_label=0)
axes[1].imshow(color2)
axes[1].set_title('Expanded labels')
```

```python
for a in axes:
    a.axis('off')
fig.tight_layout()
plt.show()
```
