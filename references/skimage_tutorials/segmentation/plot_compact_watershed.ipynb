Find Regular Segments Using Compact Watershed<br>
=============================================<br>
The watershed transform is commonly used as a starting point for many<br>
segmentation algorithms. However, without a judicious choice of seeds, it<br>
can produce very uneven fragment sizes, which can be difficult to deal with<br>
in downstream analyses.<br>
The *compact* watershed transform remedies this by favoring seeds that are<br>
close to the pixel being considered.<br>
Both algorithms are implemented in the :py:func:`skimage.morphology.watershed`<br>
function. To use the compact form, simply pass a ``compactness`` value greater<br>
than 0.<br>


```python
import numpy as np
from skimage import data, util, filters, color
from skimage.segmentation import watershed
import matplotlib.pyplot as plt
```

```python
coins = data.coins()
edges = filters.sobel(coins)
```

```python
grid = util.regular_grid(coins.shape, n_points=468)
```

```python
seeds = np.zeros(coins.shape, dtype=int)
seeds[grid] = np.arange(seeds[grid].size).reshape(seeds[grid].shape) + 1
```

```python
w0 = watershed(edges, seeds)
w1 = watershed(edges, seeds, compactness=0.01)
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2)
```

```python
ax0.imshow(color.label2rgb(w0, coins, bg_label=-1))
ax0.set_title('Classical watershed')
```

```python
ax1.imshow(color.label2rgb(w1, coins, bg_label=-1))
ax1.set_title('Compact watershed')
```

```python
plt.show()
```
