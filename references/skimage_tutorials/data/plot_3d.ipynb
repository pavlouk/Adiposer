
<br>
==========================================<br>
Datasets with 3 or more spatial dimensions<br>
==========================================<br>
Most scikit-image functions are compatible with 3D datasets, i.e., images with<br>
3 spatial dimensions (to be distinguished from 2D multichannel images, which <br>
are also arrays with<br>
three axes). :func:`skimage.data.cells3d` returns a 3D fluorescence microscopy<br>
image of cells. The returned dataset is a 3D multichannel image with dimensions<br>
provided in ``(z, c, y, x)`` order. Channel 0 contains cell membranes, while channel<br>
1 contains nuclei.<br>
The example below shows how to explore this dataset. This 3D image can be used<br>
to test the various functions of scikit-image.<br>


```python
from skimage import data
import plotly
import plotly.express as px
```

```python
img = data.cells3d()[20:]
fig = px.imshow(img, facet_col=1, animation_frame=0,
                binary_string=True, binary_format='jpg')
fig.layout.annotations[0]['text'] = 'Cell membranes'
fig.layout.annotations[1]['text'] = 'Nuclei'
plotly.io.show(fig)
```
