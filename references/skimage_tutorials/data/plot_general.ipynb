General-purpose images<br>
======================<br>
The title of each image indicates the name of the function.<br>


```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
from skimage import data
```

```python
matplotlib.rcParams['font.size'] = 18
```

```python
images = ('astronaut',
          'binary_blobs',
          'brick',
          'colorwheel',
          'camera',
          'cat',
          'checkerboard',
          'clock',
          'coffee',
          'coins',
          'eagle',
          'grass',
          'gravel',
          'horse',
          'logo',
          'page',
          'text',
          'rocket',
          )
```

```python
for name in images:
    caller = getattr(data, name)
    image = caller()
    plt.figure()
    plt.title(name)
    if image.ndim == 2:
        plt.imshow(image, cmap=plt.cm.gray)
    else:
        plt.imshow(image)
```

```python
plt.show()
```
