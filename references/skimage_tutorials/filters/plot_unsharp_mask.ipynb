Unsharp masking<br>
===============<br>
Unsharp masking is a linear image processing technique which sharpens<br>
the image. The sharp details are identified as a difference between<br>
the original image and its blurred version. These details are then scaled,<br>
and added back to the original image:<br>
    enhanced image = original + amount * (original - blurred)<br>
The blurring step could use any image filter method, e.g. median filter,<br>
but traditionally a gaussian filter is used. The radius parameter in the<br>
unsharp masking filter refers to the sigma parameter of the gaussian filter.<br>
This example shows the effect of different radius and amount parameters.<br>


```python
from skimage import data
from skimage.filters import unsharp_mask
import matplotlib.pyplot as plt
```

```python
image = data.moon()
result_1 = unsharp_mask(image, radius=1, amount=1)
result_2 = unsharp_mask(image, radius=5, amount=2)
result_3 = unsharp_mask(image, radius=20, amount=1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,
                         sharex=True, sharey=True, figsize=(10, 10))
ax = axes.ravel()
```

```python
ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title('Original image')
ax[1].imshow(result_1, cmap=plt.cm.gray)
ax[1].set_title('Enhanced image, radius=1, amount=1.0')
ax[2].imshow(result_2, cmap=plt.cm.gray)
ax[2].set_title('Enhanced image, radius=5, amount=2.0')
ax[3].imshow(result_3, cmap=plt.cm.gray)
ax[3].set_title('Enhanced image, radius=20, amount=1.0')
```

```python
for a in ax:
    a.axis('off')
fig.tight_layout()
plt.show()
```
