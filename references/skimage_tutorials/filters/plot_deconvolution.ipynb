Image Deconvolution<br>
=====================<br>
In this example, we deconvolve an image using Richardson-Lucy<br>
deconvolution algorithm ([1]_, [2]_).<br>
The algorithm is based on a PSF (Point Spread Function),<br>
where PSF is described as the impulse response of the<br>
optical system. The blurred image is sharpened through a number of<br>
iterations, which needs to be hand-tuned.<br>
.. [1] William Hadley Richardson, "Bayesian-Based Iterative<br>
       Method of Image Restoration",<br>
       J. Opt. Soc. Am. A 27, 1593-1607 (1972), :DOI:`10.1364/JOSA.62.000055`<br>
.. [2] https://en.wikipedia.org/wiki/Richardson%E2%80%93Lucy_deconvolution<br>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.signal import convolve2d as conv2
```

```python
from skimage import color, data, restoration
```

```python
astro = color.rgb2gray(data.astronaut())
```

```python
psf = np.ones((5, 5)) / 25
astro = conv2(astro, psf, 'same')
# Add Noise to Image
astro_noisy = astro.copy()
astro_noisy += (np.random.poisson(lam=25, size=astro.shape) - 10) / 255.
```

Restore Image using Richardson-Lucy algorithm

```python
deconvolved_RL = restoration.richardson_lucy(astro_noisy, psf, iterations=30)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 5))
plt.gray()
```

```python
for a in (ax[0], ax[1], ax[2]):
       a.axis('off')
```

```python
ax[0].imshow(astro)
ax[0].set_title('Original Data')
```

```python
ax[1].imshow(astro_noisy)
ax[1].set_title('Noisy data')
```

```python
ax[2].imshow(deconvolved_RL, vmin=astro_noisy.min(), vmax=astro_noisy.max())
ax[2].set_title('Restoration using\nRichardson-Lucy')
```

```python
fig.subplots_adjust(wspace=0.02, hspace=0.2,
                    top=0.9, bottom=0.05, left=0, right=1)
plt.show()
```
