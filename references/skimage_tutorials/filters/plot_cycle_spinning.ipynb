Shift-invariant wavelet denoising<br>
=================================<br>
The discrete wavelet transform is not `shift-invariant`_.  Shift invariance can<br>
be achieved through an undecimated wavelet transform (also called stationary<br>
wavelet transform), at cost of increased redundancy (i.e. more wavelet<br>
coefficients than input image pixels).  An alternative way to approximate<br>
shift-invariance in the context of image denoising with the discrete wavelet<br>
transform is to use the technique known as "cycle spinning".  This involves<br>
averaging the results of the following 3-step procedure for multiple spatial<br>
shifts, n:<br>
1. (circularly) shift the signal by an amount, n<br>
2. apply denoising<br>
3. apply the inverse shift<br>
For 2D image denoising, we demonstrate here that such cycle-spinning can<br>
provide a substantial increase in quality, with much of the gain being<br>
achieved simply by averaging shifts of only n=0 and n=1 on each axis.<br>
.. _`shift-invariant`: https://en.wikipedia.org/wiki/Shift-invariant_system<br>


```python
import matplotlib.pyplot as plt
```

```python
from skimage.restoration import denoise_wavelet, cycle_spin
from skimage import data, img_as_float
from skimage.util import random_noise
from skimage.metrics import peak_signal_noise_ratio
```

```python
original = img_as_float(data.chelsea()[100:250, 50:300])
```

```python
sigma = 0.155
noisy = random_noise(original, var=sigma**2)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 4),
                       sharex=False, sharey=False)
ax = ax.ravel()
```

```python
psnr_noisy = peak_signal_noise_ratio(original, noisy)
ax[0].imshow(noisy)
ax[0].axis('off')
ax[0].set_title('Noisy\nPSNR={:0.4g}'.format(psnr_noisy))
```

Repeat denosing with different amounts of cycle spinning.  e.g.<br>
max_shift = 0 -> no cycle spinning<br>
max_shift = 1 -> shifts of (0, 1) along each axis<br>
max_shift = 3 -> shifts of (0, 1, 2, 3) along each axis<br>
etc...

```python
denoise_kwargs = dict(channel_axis=-1, convert2ycbcr=True, wavelet='db1',
                      rescale_sigma=True)
```

```python
all_psnr = []
max_shifts = [0, 1, 3, 5]
for n, s in enumerate(max_shifts):
    im_bayescs = cycle_spin(noisy, func=denoise_wavelet, max_shifts=s,
                            func_kw=denoise_kwargs, channel_axis=-1)
    ax[n+1].imshow(im_bayescs)
    ax[n+1].axis('off')
    psnr = peak_signal_noise_ratio(original, im_bayescs)
    if s == 0:
        ax[n+1].set_title(
            "Denoised: no cycle shifts\nPSNR={:0.4g}".format(psnr))
    else:
        ax[n+1].set_title(
            "Denoised: {0}x{0} shifts\nPSNR={1:0.4g}".format(s+1, psnr))
    all_psnr.append(psnr)
```

plot PSNR as a function of the degree of cycle shifting

```python
ax[5].plot(max_shifts, all_psnr, 'k.-')
ax[5].set_ylabel('PSNR (dB)')
ax[5].set_xlabel('max cycle shift along each axis')
ax[5].grid(True)
plt.subplots_adjust(wspace=0.35, hspace=0.35)
```

Annotate with a cyan arrow on the 6x6 case vs. no cycle shift case to<br>
illustrate a region with reduced block-like artifact with cycle shifting

```python
arrowprops = dict(arrowstyle="simple,tail_width=0.1,head_width=0.5",
                  connectionstyle="arc3",
                  color='c')
for i in [1, 4]:
    ax[i].annotate("", xy=(101, 39), xycoords='data',
                   xytext=(70, 70), textcoords='data',
                   arrowprops=arrowprops)
```
