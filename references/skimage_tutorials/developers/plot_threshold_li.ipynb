
<br>
===============<br>
Li thresholding<br>
===============<br>
In 1993, Li and Lee proposed a new criterion for finding the "optimal"<br>
threshold to distinguish between the background and foreground of an image<br>
[1]_. They proposed that minimizing the *cross-entropy* between the foreground<br>
and the foreground mean, and the background and the background mean, would give<br>
the best threshold in most situations.<br>
Until 1998, though, the way to find this threshold was by trying all possible<br>
thresholds and then choosing the one with the smallest cross-entropy. At that<br>
point, Li and Tam implemented a new, iterative method to more quickly find the<br>
optimum point by using the slope of the cross-entropy [2]_. This is the method<br>
implemented in scikit-image's :func:`skimage.filters.threshold_li`.<br>
Here, we demonstrate the cross-entropy and its optimization by Li's iterative<br>
method. Note that we are using the private function `_cross_entropy`, which<br>
should not be used in production code, as it could change.<br>
.. [1] Li C.H. and Lee C.K. (1993) "Minimum Cross Entropy Thresholding"<br>
       Pattern Recognition, 26(4): 617-625<br>
       :DOI:`10.1016/0031-3203(93)90115-D`<br>
.. [2] Li C.H. and Tam P.K.S. (1998) "An Iterative Algorithm for Minimum<br>
       Cross Entropy Thresholding" Pattern Recognition Letters, 18(8): 771-776<br>
       :DOI:`10.1016/S0167-8655(98)00057-9`<br>


```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
from skimage import filters
from skimage.filters.thresholding import _cross_entropy
```

```python
cell = data.cell()
camera = data.camera()
```

#############################################################################<br>
First, we let's plot the cross entropy for the :func:`skimage.data.camera`<br>
image at all possible thresholds.

```python
thresholds = np.arange(np.min(camera) + 1.5, np.max(camera) - 1.5)
entropies = [_cross_entropy(camera, t) for t in thresholds]
```

```python
optimal_camera_threshold = thresholds[np.argmin(entropies)]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(8, 3))
```

```python
ax[0].imshow(camera, cmap='gray')
ax[0].set_title('image')
ax[0].set_axis_off()
```

```python
ax[1].imshow(camera > optimal_camera_threshold, cmap='gray')
ax[1].set_title('thresholded')
ax[1].set_axis_off()
```

```python
ax[2].plot(thresholds, entropies)
ax[2].set_xlabel('thresholds')
ax[2].set_ylabel('cross-entropy')
ax[2].vlines(optimal_camera_threshold,
             ymin=np.min(entropies) - 0.05 * np.ptp(entropies),
             ymax=np.max(entropies) - 0.05 * np.ptp(entropies))
ax[2].set_title('optimal threshold')
```

```python
fig.tight_layout()
```

```python
print('The brute force optimal threshold is:', optimal_camera_threshold)
print('The computed optimal threshold is:', filters.threshold_li(camera))
```

```python
plt.show()
```

#############################################################################<br>
Next, let's use the ``iter_callback`` feature of ``threshold_li`` to examine<br>
the optimization process as it happens.

```python
iter_thresholds = []
```

```python
optimal_threshold = filters.threshold_li(camera,
                                         iter_callback=iter_thresholds.append)
iter_entropies = [_cross_entropy(camera, t) for t in iter_thresholds]
```

```python
print('Only', len(iter_thresholds), 'thresholds examined.')
```

```python
fig, ax = plt.subplots()
```

```python
ax.plot(thresholds, entropies, label='all threshold entropies')
ax.plot(iter_thresholds, iter_entropies, label='optimization path')
ax.scatter(iter_thresholds, iter_entropies, c='C1')
ax.legend(loc='upper right')
```

```python
plt.show()
```

#############################################################################<br>

in some images, the cross-entropy is not *convex*, meaning having a single<br>
optimum. In this case, gradient descent could yield a threshold that is not<br>
optimal. In this example, we see how a bad initial guess for the optimization<br>
results in a poor threshold selection.

```python
iter_thresholds2 = []
```

```python
opt_threshold2 = filters.threshold_li(cell, initial_guess=64,
                                      iter_callback=iter_thresholds2.append)
```

```python
thresholds2 = np.arange(np.min(cell) + 1.5, np.max(cell) - 1.5)
entropies2 = [_cross_entropy(cell, t) for t in thresholds]
iter_entropies2 = [_cross_entropy(cell, t) for t in iter_thresholds2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(8, 3))
```

```python
ax[0].imshow(cell, cmap='magma')
ax[0].set_title('image')
ax[0].set_axis_off()
```

```python
ax[1].imshow(cell > opt_threshold2, cmap='gray')
ax[1].set_title('thresholded')
ax[1].set_axis_off()
```

```python
ax[2].plot(thresholds2, entropies2, label='all threshold entropies')
ax[2].plot(iter_thresholds2, iter_entropies2, label='optimization path')
ax[2].scatter(iter_thresholds2, iter_entropies2, c='C1')
ax[2].legend(loc='upper right')
```

In this image, amazingly, the *default* initial guess, the mean image value,<br>
actually lies *right* on top of the peak between the two "valleys" of the<br>
objective function. Without supplying an initial guess, Li's thresholding<br>
method does nothing at all!

#############################################################################<br>
To see what is going on, let's define a function, ``li_gradient``, that<br>
replicates the inner loop of the Li method and returns the *change* from the<br>
current threshold value to the next one. When this gradient is 0, we are at<br>
a so-called *stationary point* and Li returns this value. When it is<br>
negative, the next threshold guess will be lower, and when it is positive,<br>
the next guess will be higher.<br>
<br>
In the plot below, we show the cross-entropy and the Li update path when the<br>
initial guess is on the *right* side of that entropy peak. We overlay the<br>
threshold update gradient, marking the 0 gradient line and the default<br>
initial guess by ``threshold_li``.

```python
def li_gradient(image, t):
    """Find the threshold update at a given threshold."""
    foreground = image > t
    mean_fore = np.mean(image[foreground])
    mean_back = np.mean(image[~foreground])
    t_next = ((mean_back - mean_fore) /
              (np.log(mean_back) - np.log(mean_fore)))
    dt = t_next - t
    return dt
```

```python
iter_thresholds4 = []
opt_threshold4 = filters.threshold_li(cell, initial_guess=68,
                                      iter_callback=iter_thresholds4.append)
iter_entropies4 = [_cross_entropy(cell, t) for t in iter_thresholds4]
print(len(iter_thresholds4), 'examined, optimum:', opt_threshold4)
```

```python
gradients = [li_gradient(cell, t) for t in thresholds2]
```

```python
fig, ax1 = plt.subplots()
ax1.plot(thresholds2, entropies2)
ax1.plot(iter_thresholds4, iter_entropies4)
ax1.scatter(iter_thresholds4, iter_entropies4, c='C1')
ax1.set_xlabel('threshold')
ax1.set_ylabel('cross entropy', color='C0')
ax1.tick_params(axis='y', labelcolor='C0')
```

```python
ax2 = ax1.twinx()
ax2.plot(thresholds2, gradients, c='C3')
ax2.hlines([0], xmin=thresholds2[0], xmax=thresholds2[-1],
           colors='gray', linestyles='dashed')
ax2.vlines(np.mean(cell), ymin=np.min(gradients), ymax=np.max(gradients),
           colors='gray', linestyles='dashed')
ax2.set_ylabel(r'$\Delta$(threshold)', color='C3')
ax2.tick_params(axis='y', labelcolor='C3')
```

```python
fig.tight_layout()
```

```python
plt.show()
```

#############################################################################<br>
In addition to allowing users to provide a number as an initial guess,<br>
:func:`skimage.filters.threshold_li` can receive a function that makes a<br>
guess from the image intensities, just like :func:`numpy.mean` does by<br>
default. This might be a good option when many images with different ranges<br>
need to be processed.

```python
def quantile_95(image):
    # you can use np.quantile(image, 0.95) if you have NumPy>=1.15
    return np.percentile(image, 95)
```

```python
iter_thresholds5 = []
opt_threshold5 = filters.threshold_li(cell, initial_guess=quantile_95,
                                      iter_callback=iter_thresholds5.append)
iter_entropies5 = [_cross_entropy(cell, t) for t in iter_thresholds5]
print(len(iter_thresholds5), 'examined, optimum:', opt_threshold5)
```

```python
fig, ax1 = plt.subplots()
ax1.plot(thresholds2, entropies2)
ax1.plot(iter_thresholds5, iter_entropies5)
ax1.scatter(iter_thresholds5, iter_entropies5, c='C1')
ax1.set_xlabel('threshold')
ax1.set_ylabel('cross entropy', color='C0')
ax1.tick_params(axis='y', labelcolor='C0')
```
