Haar-like feature descriptor<br>
============================<br>
Haar-like features are simple digital image features that were introduced in a<br>
real-time face detector [1]_. These features can be efficiently computed on any<br>
scale in constant time, using an integral image [1]_. After that, a small<br>
number of critical features is selected from this large set of potential<br>
features (e.g., using AdaBoost learning algorithm as in [1]_). The following<br>
example will show the mechanism to build this family of descriptors.<br>
References<br>
----------<br>
.. [1] Viola, Paul, and Michael J. Jones. "Robust real-time face<br>
       detection." International journal of computer vision 57.2<br>
       (2004): 137-154.<br>
       https://www.merl.com/publications/docs/TR2004-043.pdf<br>
       :DOI:`10.1109/CVPR.2001.990517`<br>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage.feature import haar_like_feature_coord
from skimage.feature import draw_haar_like_feature
```

#############################################################################<br>
Different types of Haar-like feature descriptors<br>
#############################################################################<br>
The Haar-like feature descriptors come into 5 different types as illustrated<br>
in the figure below. The value of the descriptor is equal to the difference<br>
between the sum of intensity values in the green and the red one.

```python
images = [np.zeros((2, 2)), np.zeros((2, 2)),
          np.zeros((3, 3)), np.zeros((3, 3)),
          np.zeros((2, 2))]
```

```python
feature_types = ['type-2-x', 'type-2-y',
                 'type-3-x', 'type-3-y',
                 'type-4']
```

```python
fig, axs = plt.subplots(3, 2)
for ax, img, feat_t in zip(np.ravel(axs), images, feature_types):
    coord, _ = haar_like_feature_coord(img.shape[0], img.shape[1], feat_t)
    haar_feature = draw_haar_like_feature(img, 0, 0,
                                          img.shape[0],
                                          img.shape[1],
                                          coord,
                                          max_n_features=1,
                                          random_state=0)
    ax.imshow(haar_feature)
    ax.set_title(feat_t)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig.suptitle('The different Haar-like feature descriptors')
plt.axis('off')
plt.show()
```

#############################################################################<br>

intensity values in the green rectangle and the red one. The red area is<br>
subtracted to the sum of the pixel intensities of the green In practice, the<br>
Haar-like features will be placed in all possible location of an image and a<br>
