Gabors / Primary Visual Cortex "Simple Cells" from an Image<br>
============================================================<br>
How to build a (bio-plausible) *sparse* dictionary (or 'codebook', or<br>
'filterbank') for e.g. image classification without any fancy math and<br>
with just standard python scientific libraries?<br>
Please find below a short answer ;-)<br>
This simple example shows how to get Gabor-like filters [1]_ using just<br>
a simple image. In our example, we use a photograph of the astronaut Eileen<br>
Collins. Gabor filters are good approximations of the "Simple Cells" [2]_<br>
receptive fields [3]_ found in the mammalian primary visual cortex (V1)<br>
(for details, see e.g. the Nobel-prize winning work of Hubel & Wiesel done<br>
in the 60s [4]_ [5]_).<br>
Here we use McQueen's 'kmeans' algorithm [6]_, as a simple biologically<br>
plausible hebbian-like learning rule and we apply it (a) to patches of<br>
the original image (retinal projection), and (b) to patches of an<br>
LGN-like [7]_ image using a simple difference of gaussians (DoG)<br>
approximation.<br>
Enjoy ;-) And keep in mind that getting Gabors on natural image patches<br>
is not rocket science.<br>
.. [1] https://en.wikipedia.org/wiki/Gabor_filter<br>
.. [2] https://en.wikipedia.org/wiki/Simple_cell<br>
.. [3] https://en.wikipedia.org/wiki/Receptive_field<br>
.. [4] D. H. Hubel and T. N., Wiesel Receptive Fields of Single Neurones<br>
       in the Cat's Striate Cortex, J. Physiol. pp. 574-591 (148) 1959<br>
.. [5] D. H. Hubel and T. N., Wiesel Receptive Fields, Binocular<br>
       Interaction, and Functional Architecture in the Cat's Visual Cortex,<br>
       J. Physiol. 160 pp.  106-154 1962<br>
.. [6] https://en.wikipedia.org/wiki/K-means_clustering<br>
.. [7] https://en.wikipedia.org/wiki/Lateral_geniculate_nucleus<br>


```python
import numpy as np
from scipy.cluster.vq import kmeans2
from scipy import ndimage as ndi
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage import color
from skimage.util.shape import view_as_windows
from skimage.util import montage
```

```python
np.random.seed(42)
```

```python
patch_shape = 8, 8
n_filters = 49
```

```python
astro = color.rgb2gray(data.astronaut())
```

-- filterbank1 on original image

```python
patches1 = view_as_windows(astro, patch_shape)
patches1 = patches1.reshape(-1, patch_shape[0] * patch_shape[1])[::8]
fb1, _ = kmeans2(patches1, n_filters, minit='points')
fb1 = fb1.reshape((-1,) + patch_shape)
fb1_montage = montage(fb1, rescale_intensity=True)
```

-- filterbank2 LGN-like image

```python
astro_dog = ndi.gaussian_filter(astro, .5) - ndi.gaussian_filter(astro, 1)
patches2 = view_as_windows(astro_dog, patch_shape)
patches2 = patches2.reshape(-1, patch_shape[0] * patch_shape[1])[::8]
fb2, _ = kmeans2(patches2, n_filters, minit='points')
fb2 = fb2.reshape((-1,) + patch_shape)
fb2_montage = montage(fb2, rescale_intensity=True)
```

-- plotting

```python
fig, axes = plt.subplots(2, 2, figsize=(7, 6))
ax = axes.ravel()
```

```python
ax[0].imshow(astro, cmap=plt.cm.gray)
ax[0].set_title("Image (original)")
```

```python
ax[1].imshow(fb1_montage, cmap=plt.cm.gray)
ax[1].set_title("K-means filterbank (codebook)\non original image")
```

```python
ax[2].imshow(astro_dog, cmap=plt.cm.gray)
ax[2].set_title("Image (LGN-like DoG)")
```

```python
ax[3].imshow(fb2_montage, cmap=plt.cm.gray)
ax[3].set_title("K-means filterbank (codebook)\non LGN-like DoG image")
```

```python
for a in ax.ravel():
    a.axis('off')
```

```python
fig.tight_layout()
plt.show()
```
