===============================<br>
Filling holes and finding peaks<br>
===============================<br>
We fill holes (i.e. isolated, dark spots) in an image using morphological<br>
reconstruction by erosion. Erosion expands the minimal values of the seed image<br>
until it encounters a mask image. Thus, the seed image and mask image represent<br>
the maximum and minimum possible values of the reconstructed image.<br>
We start with an image containing both peaks and holes:<br>


```python
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage.exposure import rescale_intensity
```

```python
image = data.moon()
# Rescale image intensity so that we can see dim features.
image = rescale_intensity(image, in_range=(50, 200))
```

####################################################################<br>
Now we need to create the seed image, where the minima represent the<br>
starting points for erosion.  To fill holes, we initialize the seed image<br>
to the maximum value of the original image. Along the borders, however, we<br>
use the original values of the image. These border pixels will be the<br>
starting points for the erosion process. We then limit the erosion by<br>
setting the mask to the values of the original image.

```python
import numpy as np
from skimage.morphology import reconstruction
```

```python
seed = np.copy(image)
seed[1:-1, 1:-1] = image.max()
mask = image
```

```python
filled = reconstruction(seed, mask, method='erosion')
```

####################################################################<br>
As shown above, eroding inward from the edges removes holes, since (by<br>
definition) holes are surrounded by pixels of brighter value. Finally, we<br>
can isolate the dark regions by subtracting the reconstructed image from<br>
the original image.<br>
<br>
Alternatively, we can find bright spots in an image using morphological<br>
reconstruction by dilation. Dilation is the inverse of erosion and expands<br>
the *maximal* values of the seed image until it encounters a mask image.<br>
Since this is an inverse operation, we initialize the seed image to the<br>
minimum image intensity instead of the maximum. The remainder of the<br>
process is the same.

```python
seed = np.copy(image)
seed[1:-1, 1:-1] = image.min()
rec = reconstruction(seed, mask, method='dilation')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(5, 4), sharex=True, sharey=True)
ax = ax.ravel()
```

```python
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original image')
ax[0].axis('off')
```

```python
ax[1].imshow(filled, cmap='gray')
ax[1].set_title('after filling holes')
ax[1].axis('off')
```

```python
ax[2].imshow(image-filled, cmap='gray')
ax[2].set_title('holes')
ax[2].axis('off')
```

```python
ax[3].imshow(image-rec, cmap='gray')
ax[3].set_title('peaks')
ax[3].axis('off')
plt.show()
```
