=================<br>
Template Matching<br>
=================<br>
We use template matching to identify the occurrence of an image patch<br>
(in this case, a sub-image centered on a single coin). Here, we<br>
return a single match (the exact same coin), so the maximum value in the<br>
``match_template`` result corresponds to the coin location. The other coins<br>
look similar, and thus have local maxima; if you expect multiple matches, you<br>
should use a proper peak-finding function.<br>
The ``match_template`` function uses fast, normalized cross-correlation [1]_<br>
to find instances of the template in the image. Note that the peaks in the<br>
output of ``match_template`` correspond to the origin (i.e. top-left corner) of<br>
the template.<br>
.. [1] J. P. Lewis, "Fast Normalized Cross-Correlation", Industrial Light and<br>
       Magic.<br>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage.feature import match_template
```

```python
image = data.coins()
coin = image[170:220, 75:130]
```

```python
result = match_template(image, coin)
ij = np.unravel_index(np.argmax(result), result.shape)
x, y = ij[::-1]
```

```python
fig = plt.figure(figsize=(8, 3))
ax1 = plt.subplot(1, 3, 1)
ax2 = plt.subplot(1, 3, 2)
ax3 = plt.subplot(1, 3, 3, sharex=ax2, sharey=ax2)
```

```python
ax1.imshow(coin, cmap=plt.cm.gray)
ax1.set_axis_off()
ax1.set_title('template')
```

```python
ax2.imshow(image, cmap=plt.cm.gray)
ax2.set_axis_off()
ax2.set_title('image')
# highlight matched region
hcoin, wcoin = coin.shape
rect = plt.Rectangle((x, y), wcoin, hcoin, edgecolor='r', facecolor='none')
ax2.add_patch(rect)
```

```python
ax3.imshow(result)
ax3.set_axis_off()
ax3.set_title('`match_template`\nresult')
# highlight matched region
ax3.autoscale(False)
ax3.plot(x, y, 'o', markeredgecolor='r', markerfacecolor='none', markersize=10)
```

```python
plt.show()
```
