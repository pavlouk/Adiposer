
<br>
===========<br>
Convex Hull<br>
===========<br>
The convex hull of a binary image is the set of pixels included in the<br>
smallest convex polygon that surround all white pixels in the input.<br>
A good overview of the algorithm is given on `Steve Eddin's blog<br>
<https://blogs.mathworks.com/steve/2011/10/04/binary-image-convex-hull-algorithm-notes/>`__.<br>


```python
import matplotlib.pyplot as plt
```

```python
from skimage.morphology import convex_hull_image
from skimage import data, img_as_float
from skimage.util import invert
```

The original image is inverted as the object must be white.

```python
image = invert(data.horse())
```

```python
chull = convex_hull_image(image)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()
```

```python
ax[0].set_title('Original picture')
ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_axis_off()
```

```python
ax[1].set_title('Transformed picture')
ax[1].imshow(chull, cmap=plt.cm.gray)
ax[1].set_axis_off()
```

```python
plt.tight_layout()
plt.show()
```

####################################################################<br>
We prepare a second plot to show the difference.<br>


```python
chull_diff = img_as_float(chull.copy())
chull_diff[image] = 2
```

```python
fig, ax = plt.subplots()
ax.imshow(chull_diff, cmap=plt.cm.gray)
ax.set_title('Difference')
plt.show()
```
