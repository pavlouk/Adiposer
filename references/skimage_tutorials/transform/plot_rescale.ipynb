Rescale, resize, and downscale<br>
==============================<br>
`Rescale` operation resizes an image by a given scaling factor. The scaling<br>
factor can either be a single floating point value, or multiple values - one<br>
along each axis.<br>
`Resize` serves the same purpose, but allows to specify an output image shape<br>
instead of a scaling factor.<br>
Note that when down-sampling an image, `resize` and `rescale` should perform<br>
Gaussian smoothing to avoid aliasing artifacts. See the `anti_aliasing` and<br>
`anti_aliasing_sigma` arguments to these functions.<br>
`Downscale` serves the purpose of down-sampling an n-dimensional image by<br>
integer factors using the local mean on the elements of each block of the size<br>
factors given as a parameter to the function.<br>


```python
import matplotlib.pyplot as plt
```

```python
from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
image = color.rgb2gray(data.astronaut())
```

```python
image_rescaled = rescale(image, 0.25, anti_aliasing=False)
image_resized = resize(image, (image.shape[0] // 4, image.shape[1] // 4),
                       anti_aliasing=True)
image_downscaled = downscale_local_mean(image, (4, 3))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
ax = axes.ravel()
```

```python
ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")
```

```python
ax[1].imshow(image_rescaled, cmap='gray')
ax[1].set_title("Rescaled image (aliasing)")
```

```python
ax[2].imshow(image_resized, cmap='gray')
ax[2].set_title("Resized image (no aliasing)")
```

```python
ax[3].imshow(image_downscaled, cmap='gray')
ax[3].set_title("Downscaled image (no aliasing)")
```

```python
ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```
